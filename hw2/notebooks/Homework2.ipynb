## Importación de flujos en Kestra

Una vez que el servicio de Kestra esté en funcionamiento, siga estos pasos para importar los flujos de trabajo:

1. **Acceso a la Interfaz Web de Kestra**  
   Abra su navegador web y diríjase a la URL donde se está ejecutando Kestra (por ejemplo, `http://localhost:8080`).

2. **Importación de Flujos**  
   En la interfaz de Kestra, busque la opción para **importar flujos**. Luego, navegue hasta la carpeta `flow` del repositorio y seleccione los archivos de definición de los flujos que desee importar.

    Si prefieres automatizar el proceso, puedes importar flujos mediante la API de Kestra. Para ello, envía una solicitud POST a la API de importación de flujos, especificando el archivo o los archivos a importar.  
         
    Por ejemplo, si tienes un archivo `flows/my_flow.yaml`, podrías utilizar un comando `curl` similar al siguiente:
     
 ```bash
   curl -X POST http://localhost:8080/api/v1/flows/import -F fileUpload=@flows/my_flow.yaml
 ```
   

4. **Verificación y Ejecución**  
   Una vez importados, revise que los flujos se hayan cargado correctamente en la plataforma. Desde la interfaz, podrá visualizarlos, editarlos y ejecutarlos según sea necesario.

Con estos pasos, sus flujos de trabajo quedarán disponibles y listos para ser gestionados a través de Kestra.
