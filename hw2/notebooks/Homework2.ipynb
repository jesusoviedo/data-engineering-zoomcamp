**timezone: "America/New_York"** -> **[kestra trigger docs](https://kestra.io/plugins/core/triggers/io.kestra.plugin.core.trigger.schedule#timezone)**

```yaml
triggers:
  - id: my_schedule
    type: io.kestra.plugin.core.trigger.Schedule
    cron: "0 9 * * *"
    timezone: "America/New_York"
```
